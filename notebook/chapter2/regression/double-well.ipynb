```python
import numpy as np
import matplotlib.pyplot as plt

import os
os.makedirs("figs", exist_ok=True)
# Define the double-well potential and its derivative
def U(x):
    return 0.25 * x**4 - 0.5 * x**2

# Define a safe energy function
def compute_energy(x, x_dot):
    return 0.25 * x**4 - 0.5 * x**2 + 0.5 * x_dot**2

# Generate a grid for x and x_dot
x = np.linspace(-2, 2, 500)
x_dot = np.linspace(-2, 2, 500)
X, X_dot = np.meshgrid(x, x_dot)

# Calculate the total energy E = U(x) + 0.5 * x_dot^2
E = compute_energy(X, X_dot)

# Mask out invalid or extremely large values for stability
E[E > 5] = np.nan  # Exclude values outside the energy range of interest

# Energy levels for isolines
energy_levels = np.linspace(-1, 0.5, 6)

# Plot phase portrait isolines
plt.figure(figsize=(8, 6))
contours = plt.contour(X, X_dot, E, levels=energy_levels, colors='blue', linestyles='-')
plt.clabel(contours, inline=True, fontsize=8, fmt="E=%.1f")

# Add labels, grid, and title
plt.title("Phase Portrait for Double-Well Potential")
plt.xlabel("$x$ (Position)")
plt.ylabel("$\dot{x}$ (Velocity)")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)
fig = plt.gcf()
fig.tight_layout()
fig_path = os.path.join('figs', 'double-well-phase.png')
fig.savefig(fig_path, dpi=200, bbox_inches='tight')
print('Saved figure:', fig_path)
plt.show()
```
